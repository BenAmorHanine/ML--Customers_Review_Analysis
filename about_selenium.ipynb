# Important Considerations

Web Drivers: Selenium requires a driver to interface with the chosen browser. For Firefox, you need geckodriver, and for Chrome, you need chromedriver. Ensure these drivers are installed and added to your system PATH.

Browser Compatibility: Selenium supports multiple browsers like Firefox, Chrome, Safari, and Edge. Make sure to download the appropriate driver for the browser you intend to use.

Headless Mode: For running tests without opening a browser window, you can use headless mode. This is useful for running scripts in environments without a graphical interface.

By following these steps, you can start automating web tasks using Selenium WebDriver in Python efficiently

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import time
import pandas as pd

# Chemin absolu vers chromedriver.exe
chromedriver_path = r"E:\ML Cours\Project\ML--Customers_Review_Analysis\chromedriver-win64\chromedriver.exe"
service = Service(chromedriver_path)

# Options Chrome (optionnel : headless pour ne pas ouvrir la fenêtre)
options = webdriver.ChromeOptions()
# options.add_argument("--headless")

# Lancement du navigateur
driver = webdriver.Chrome(service=service, options=options)

# URL du produit (Tunisianet, Redmi Note 13)
url = "https://www.tunisianet.com.tn/smartphone/71310-smartphone-xiaomi-redmi-note-13-8-go-256-go-noir.html"
driver.get(url)

# Attendre que la page et les avis se chargent
time.sleep(5)

# Récupération des avis
reviews = driver.find_elements("class name", "review-item")
data = []

for review in reviews:
    try:
        name = review.find_element("class name", "reviewer-name").text
    except:
        name = "Inconnu"

    try:
        date = review.find_element("class name", "review-date").text
    except:
        date = "Inconnue"

    try:
        stars = review.find_elements("css selector", ".star_content i.fa-star")
        rating = len(stars)
    except:
        rating = None

    try:
        comment = review.find_element("class name", "review-text").text
    except:
        comment = ""

    data.append({
        "name": name,
        "date": date,
        "rating": rating,
        "comment": comment
    })

# Convertir en DataFrame et sauvegarder CSV
df = pd.DataFrame(data)
df.to_csv(r"E:\ML Cours\Project\ML--Customers_Review_Analysis\reviews_redmi_note_13.csv", index=False)

print("✅ CSV sauvegardé avec succès !")
print(df.head())

# Fermer le navigateur
driver.quit()
```
