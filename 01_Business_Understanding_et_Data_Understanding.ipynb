Confirmation of the Business Hotel's Identity
Located at 4 Rue Ibn Ennadim in central Tunis, near major business districts and the airport, this hotel is a distinct entity with 39 rooms, 3 suites, meeting facilities, a rooftop bar (Luna Sky), and a pub (B2B). It targets business clients but welcomes all travelers, with amenities like high-speed WiFi, event spaces, and live entertainment. Multiple platforms confirm its existence and operations since at least 2015, with no indications of it being a generic term—it's a branded hotel under independent management.

Officially branded as Business Hôtel Tunis or Business Hotel Tunis, this 4-star property is situated at 4 Rue Ibn Ennadim in the Montplaisir district of central Tunis, Tunisia, approximately 10 km from Tunis-Carthage International Airport (about a 10-15 minute drive). Established to cater to business travelers, it features 39 standard rooms and 3 suites, all equipped with modern amenities like air-conditioned spaces, flat-screen TVs, safety deposit boxes, minibars, and coffee machines. The hotel's design emphasizes professional needs, including two dedicated meeting rooms: the Didon Room (capacity for 18 people) and the Casa Luna Room (capacity for 25), both supporting videoconferencing for hybrid events. Leisure elements include the Luna Sky Rooftop on the 6th floor, offering panoramic city views, daily live music, and Tunisian gastronomy, alongside the B2B Pub on the first floor for casual drinks and entertainment. Additional services encompass 24-hour room service, a business center with fax and photocopying, concierge assistance, dry cleaning, and on-site parking. The restaurant focuses on international and local cuisine, with a daily continental breakfast praised for its variety, including fresh Tunisian specialties. Spaces like the rooftop and pub can be privatized for events, making it versatile for corporate gatherings, family stays, or tourism. Its central location places it within walking distance of landmarks like Habib Bourguiba Avenue (9 minutes away), Belveder Parc (14 minutes), and cultural sites such as the Al-Zaytuna Mosque, while being near business hubs and public transport (e.g., Mohamed V bus station 200 meters away).
